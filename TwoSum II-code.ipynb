```
# tried this with two pointers, seemed like the cleanest way
# since the array is already sorted we dont need a hashmap

def twoSum(numbers, target):
    l = 0
    r = len(numbers) - 1

    while l < r:
        s = numbers[l] + numbers[r]
        if s == target:
            return [l + 1, r + 1]
        elif s < target:
            l += 1
        else:
            r -= 1

# quick checks
print(twoSum([2, 7, 11, 15], 9))   # [1, 2]
print(twoSum([2, 3, 4], 6))        # [1, 3]
print(twoSum([-1, 0], -1))         # [1, 2]
```
