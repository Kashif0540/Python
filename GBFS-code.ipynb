```
graph = {
    'S': {'A': 3, 'C': 2, 'D': 2},
    'A': {},
    'B': {'E': 2},
    'C': {'F': 1},
    'D': {'B': 3, 'G': 8},
    'E': {'G': 2},
    'F': {'E': 0, 'G': 4},
    'G': {}
}

Heuristics = {
    'S': 20,
    'A': 18,
    'C': 16,
    'D': 14,
    'B': 10,
    'E': 9,
    'F': 10,
    'G': 0
}

def GBFS(start, goal, graph, heuristic):
    queue = [(heuristic[start], 0, [start])]
    visited = []

    while queue:
        # ⭐ MAIN DIFFERENCE LINE (GBFS LOGIC)
        queue.sort()
        # GBFS: sorts only by heuristic h(n), ignores actual path cost g(n)

        h_value, cost, path = queue.pop(0)  # pick node with smallest h(n)
        node = path[-1]

        if node in visited:
            continue

        visited.append(node)

        if node == goal:
            return path, cost

        for child in graph[node]:
            newCost = cost + graph[node][child]
            newPath = path + [child]

            queue.append((heuristic[child], newCost, newPath))

    return None


print(GBFS('S', 'G', graph, Heuristics))




```
