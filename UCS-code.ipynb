```
graph = {
    'S': {'A': 3, 'C': 2, 'D': 2},
    'A': {},
    'B': {'E': 2},
    'C': {'F': 1},
    'D': {'B': 3, 'G': 8},
    'E': {'G': 2},
    'F': {'E': 0, 'G': 4},
    'G': {}
}

def UCS(start, goal, graph):
    queue = [(0, [start])]   # (cost g(n), path)
    visited = []

    while queue:
        #  (UCS LOGIC)
        queue.sort()
        # UCS: sorts by path cost g(n), ignores heuristic

        cost, path = queue.pop(0)  # pick node with smallest cost
        node = path[-1]

        if node in visited:
            continue

        visited.append(node)

        if node == goal:
            return path, cost

        for child in graph[node]:
            newCost = cost + graph[node][child]
            newPath = path + [child]

            queue.append((newCost, newPath))

    return None


print(UCS('S', 'G', graph))
```
