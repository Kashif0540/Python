```
graph={
    'S':{'A':3,'C':2,'D':2},
    'A':{},
    'B':{'E':2},
    'C':{'F':1},
    'D':{'B':3,'G':8},
    'E':{'G':2},
    'F':{'E':0,'G':4},
    'G':{},
}

Heuristic={
    'S':20,
    'A':18,
    'B':10,
    'C':16,
    'D':14,
    'E':9,
    'F':10,
    'G':0
}

def A_STAR(start,goal,graph,heuristic):
  queue=[(heuristic[start],0,[start])]

  visited=[]
  while queue:
    queue.sort()

    current_value,cost,path=queue.pop(0)
    node=path[-1]

    if node in visited:
      continue

    visited.append(node)

    if node==goal:
        return path,cost

    for child in graph[node]:
        newCost=cost+graph[node][child]

        total_Cost=newCost+heuristic[child]

        newPath=path+[child]
        queue.append((total_Cost,newCost,newPath))

print(A_STAR('S','G',graph,Heuristic))
```
