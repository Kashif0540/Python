```
text = "apple"
count_dict = {}

for char in text:
    # count() is a built-in function to count occurrences
    count_dict[char] = text.count(char)

print(count_dict) # Output: {'a': 1, 'p': 2, 'l': 1, 'e': 1}
```

```
# Range function: range(start, stop, step)
for i in range(1, 10, 2):
    print(i) # Prints 1, 3, 5, 7, 9
```

```
arr = [10, 5, 8, 20, 15]

mid = len(arr) // 2

first_half = sorted(arr[:mid])
second_half = arr[mid:]

result = first_half + second_half

print("First half (sorted):", first_half)
print("Second half (unchanged):", second_half)
print("Final result:", result)
```

```
with open("data.txt", "w") as f:
  f.write("2,3,4,5,6,7,8,2,3,4,5")


with open("data.txt", "r") as file:
  data=file.read().split(",")

print(data)

numbers = list(map(int,data))

freq={}
for num in numbers:
  freq[num] = freq.get(num,0)+1

non_repeating=[]
for num in numbers:
   if freq[num]==1:
     non_repeating.append(num)

if len(non_repeating) >= 2:
    print("Output:", non_repeating[1])
else:
    print("NO second non-repeating element")
```
