```
graph = {
     'A' : {'B' : 1, 'C' : 2},
     'B' : {'D' : 3, 'E' : 4},
     'C' : {'F' : 6, 'G' : 8},
     'D' : {},
     'E' : {},
     'F' : {},
     'G' : {}
}

def bfs(start,goal):
     cost=0
     path=[]
     visited=set()
     queue=[(start,cost)]

     while queue:
      current,cost=queue.pop(0)
      if current==goal:
        path.append(current)
        return path,cost
      if current not in visited:
        path.append(current)
        visited.add(current)
        for x,y in graph.get(current,{}).items():
          queue.append((x,y+cost))

print(bfs('A','G'))
```
